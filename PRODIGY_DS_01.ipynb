```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

uploaded = files.upload()

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_400028.csv', skiprows=4)

df.head()
```

```python
data_2022 = df[['Country Name', '2022']].dropna()
data_2022.columns = ['Country', 'Population']
```

```python
top10 = data_2022.sort_values(by='Population', ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top10['Country'], top10['Population'], color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Most Populous Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data_2022['Population'] / 1e6, bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Country Populations (2022)')
plt.xlabel('Population (in millions)')
plt.ylabel('Number of Countries')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
